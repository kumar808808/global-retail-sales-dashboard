```python
import duckdb
con=duckdb.connect()
con.execute(""" CREATE TABLE superstore AS
SELECT * FROM read_csv_auto(super_clean.csv) """).fetchall()
```

```python
con.execute("SELECT count(*) FROM superstore").fetchall()
```

Total_sales

```python
con.execute(""" SELECT SUM(sales) FROM superstore """).fetchall()
```

Total_profit

```python
con.execute(""" SELECT SUM(profit) FROM superstore """).fetchall()
```

Total count of orders

```python
con.execute(""" SELECT COUNT(DISTINCT "Order ID") FROM superstore""").fetchall()
```

Monthly sales trend

```python
con.execute("""
WITH monthly_sales AS(
SELECT DATE_TRUNC('month',"Order Date") AS order_month,
SUM(sales) AS current_month_sales
FROM superstore
GROUP BY 1)
SELECT order_month,
LAG(current_month_sales) OVER(ORDER BY order_month) AS previous_month_sales,
current_month_sales - LAG(current_month_sales) OVER(ORDER BY order_month) as revenue_growth
FROM monthly_sales
ORDER BY order_month
""").fetchall()
```

Which product categories generate the highest sales?

```python
con.execute("""
SELECT category,SUM(sales) as total_sales
FROM superstore
GROUP BY 1
ORDER BY 2 DESC

""").fetchall()
```

Which region generates the most revenue?

```python
con.execute("""
SELECT region,SUM(sales),SUM(profit)
FROM  superstore
GROUP BY 1
ORDER BY 2 DESC

""").fetchall()
```

Which categories are most profitable?

```python
con.execute("""
SELECT category,SUM(profit) AS total_profit
FROM superstore
GROUP BY 1
ORDER BY 2 DESC

""").fetchall()
```

Which individual products generate the most revenue?

```python
con.execute("""
SELECT
"Product Name",SUM(sales) as total_sales
FROM superstore
GROUP BY 1
ORDER BY 2 DESC limit 10

""").fetchall()
```

Which shipping mode is fastest?

```python
con.execute("""
SELECT "Ship Mode",AVG("shipping days") as avg_shipping_days
FROM superstore
GROUP BY 1
ORDER BY 2 ASC LIMIT 1
""").fetchall()
```

Which customer segment generates the most revenue?

```python
con.execute("""
SELECT segment,SUM(sales) as total_sales
 FROM superstore
 GROUP BY 1
 ORDER BY 2 DESC LIMIT 1

 """).fetchall()
```

Rank the top products by revenue within each category.

```python
con.execute("""
WITH product_revenue AS(
SELECT category,"Product Name",SUM(sales) AS total_sales,
RANK() OVER(PARTITION BY category ORDER BY SUM(sales) DESC) AS sales_rank
FROM superstore
GROUP BY 1,2
ORDER BY 1,3 DESC)
SELECT * FROM product_revenue WHERE sales_rank<=3  ORDER BY 3 desc
""").fetchall()
```
